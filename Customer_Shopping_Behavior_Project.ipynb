# Input Missing Values

```python
df.isnull().sum()
```

```python
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

# Standardizing & Organizing Columns

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df = df.rename(columns={"purchase_amount_(usd)" : "purchase_amount"})
```

```python
print(df.columns)
```

```python
# Create a column age_group

labels = ["Young Adult", "Adult", "Middle-Aged", "Senior"]
df["age_group"] = pd.qcut(df["age"], q=4, labels = labels)
```

```python
df[["age", "age_group"]].head(10)
```

```python
# Create a column purchase_frequency_days

frequency_mapping = {
    "Fortnightly": 14,
    "Weekly": 7,
    "Monthly": 30,
    "Quarterly": 90,
    "Bi-Weekly": 14,
    "Annually": 365,
    "Every 3 Months": 90
}

df["purhcase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["purhcase_frequency_days", "frequency_of_purchases"]].head(10)
```

```python
df[["discount_applied", "promo_code_used"]].head(10)
```
